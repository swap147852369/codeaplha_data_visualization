```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
##BAR plot - Top 10 most common cuisines

top_cuisines=df['Cuisines'].value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_cuisines.index,y=top_cuisines.values)
plt.title("Top 10 most common cuisines")
plt.xlabel("Number of dishes")
plt.ylabel("Cuisine")
plt.show()
```

```python
##BAR PLOT - Average rating by cuisine
avg_rating=df.groupby('Cuisines')['Overall Rating'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=avg_rating.index,y=avg_rating.values)
plt.title("Average rating by cuisine")
plt.xlabel("Cuisine")
plt.ylabel("Average rating")
plt.show()
```

```python
##Pie chart - Cuisine distribution
cuisine_distribution=df['Cuisines'].value_counts()
plt.figure(figsize=(10,6))
plt.pie(cuisine_distribution.values,labels=cuisine_distribution.index,autopct='%1.1f%%')
plt.title("Cuisine distribution")
plt.show()
```

```python
##Box plot - To detect rating outliers

plt.figure(figsize=(10,6))
sns.boxplot(x='Cuisines',y='Overall Rating',data=df)
plt.title("Rating outliers")
plt.show()
```

```python
## scatter plot - Does high food rating also have high service ratings
plt.figure(figsize=(10,6))
sns.scatterplot(x='Food Rating', y='Service Rating',data=df)
plt.title("Food Rating vs Service Rating")
plt.show()
```

```python
###CONCLUSION

### Top 10 most common cuisines
### - Japanese cuisine is the most common, suggesting high popularity.
### - This plot helps identify dominant cuisines in the dataset.

### Average rating by cuisine
### - Japanese cuisine has the highest average overall rating.
### - This indicates which cuisines are generally rated higher by users.

### Cuisine distribution
### - This pie chart visually confirms the proportion of each cuisine in the dataset.
### - It shows the relative popularity of different cuisines.

### Rating outliers
### - The box plot helps visualize the distribution of overall ratings for each cuisine and identify potential outliers.
### - Some cuisines show a wider range of ratings or potential outliers, which might warrant further investigation.

### Food Rating vs Service Rating
### - This scatter plot explores the relationship between food rating and service rating.
### - Observing the spread of points can reveal if there's a correlation between how food and service are rated.

### Final Summary

### - Japanese and French cuisines dominate in terms of presence.
### - Indian food stands out with strong food ratings.
### - Food rating is more strongly linked to overall satisfaction than service.
### - A few cuisines have excellent food but average service, which might affect customer experience.

### These insights could help restaurant owners or food delivery apps prioritize cuisine quality and service improvements.
```
